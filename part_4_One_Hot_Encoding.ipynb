```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd


df = pd.read_csv('/content/drive/MyDrive/df_after_eventtype_encoding.csv')

print(df.shape)
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

cols = ["user_id", "day_of_week", "time_of_day", "ad_platform", "ad_type",
        "target_gender", "location" ,"target_age_group" , "age_group", "user_gender", "month_name"]

for col_name in cols:
    df[col_name + "_encoded"] = encoder.fit_transform(df[col_name])

df.drop(cols , inplace= True , axis=1)
df.info()
```

###### Transform the `interests` column into multiple binary columns, where each column represents a single interest and is assigned a value of `1` if the user has that interest and `0` otherwise, enabling the data to be used for analysis and modeling.


```python
interests = [
    "fitness", "health", "fashion", "travel", "art", "technology",
    "lifestyle", "sports", "news", "photography", "gaming", "finance", "food"
]
# Split to lists
df["interests_list"] = df["interests"].str.split(", ")

# Create dummy columns manually
for interest in interests:
    df["user_interest_" + interest] = df["interests_list"].apply(lambda x: int(interest in x))

df.head()
```

###### Transform the `target_interests` column into multiple binary columns, where each column represents a single target interest and is assigned a value of `1` if it appears in the list and `0` otherwise, allowing the data to be used for analysis and modeling.

```python
interests = ["fashion",
    "photography",
    "news",
    "art",
    "health",
    "sports",
    "gaming",
    "lifestyle",
    "finance",
    "food",
    "travel",
    "fitness",
    "technology"
]

df["target_interests"] = df["target_interests"].str.split(", ")

# Create dummy columns manually
for interest in interests:
    df["target_interests_" + interest] = df["target_interests"].apply(lambda x: int(interest in x))

df.head()
```

```python
df.to_csv('/content/drive/MyDrive/df_after_encoding.csv', index=False)
```
